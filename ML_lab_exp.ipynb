```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
features = data.iloc[:,:-1]
labels = data.iloc[:,-1:]
```

```python
print(features)
```

```python
print(labels)
```

```python
from sklearn.model_selection import train_test_split
```

```python
features_train, features_test, labels_train, labels_test = train_test_split(features,labels,test_size=0.3)
```

```python
data.shape
```

```python
features_train
```

```python
features_test
```

```python
labels_train
```

```python
labels_test
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score

classifier=tree.DecisionTreeClassifier(criterion = "entropy",max_depth = 3)

cf = classifier.fit(features_train,labels_train)
```

```python
prediction = cf.predict(features_test)
```

```python
print("Accuracy = ",accuracy_score(labels_test,prediction,normalize=True))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print('Confusion matrix = ', confusion_matrix(labels_test,prediction))
```
